```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import neighbors
```

```python
advertising = pd.read_csv('https://raw.githubusercontent.com/galvin-mj/DAT_ATL_15/master/Datasets/Advertising.csv')
del advertising['Unnamed: 0']
advertising.head()
```

```python
plt.scatter(advertising['TV'],advertising['Sales'])
advertising = advertising.sort('TV')
```

```python
knn = neighbors.KNeighborsRegressor()
y_ = knn.fit(advertising['TV'].values.reshape(200,1),advertising['Sales'].values.reshape(200,1)).predict(advertising['TV'].reshape(200,1))
plt.scatter(advertising['TV'],advertising['Sales'])
plt.plot(advertising['TV'],y_,c='g')
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
y_ = knn.fit(advertising['TV'].values.reshape(200,1),advertising['Sales'].values.reshape(200,1)).predict(advertising['TV'].values.reshape(200,1))
plt.scatter(advertising['TV'].values,advertising['Sales'])
plt.plot(advertising['TV'].values,y_,c='g')
```

```python
knn = neighbors.KNeighborsRegressor(200)
y_ = knn.fit(advertising['TV'].reshape(200,1),advertising['Sales'].reshape(200,1)).predict(advertising['TV'].reshape(200,1))
plt.scatter(advertising['TV'],advertising['Sales'])
plt.plot(advertising['TV'],y_,c='g')
```

```python
knn = neighbors.KNeighborsRegressor(weights='distance')
y_ = knn.fit(advertising['TV'].reshape(200,1),advertising['Sales'].reshape(200,1)).predict(advertising['TV'].reshape(200,1))
plt.scatter(advertising['TV'],advertising['Sales'])
plt.plot(advertising['TV'],y_,c='g')
```

```python
knn = neighbors.KNeighborsRegressor(100)
y_ = knn.fit(advertising['TV'].reshape(200,1),advertising['Sales'].reshape(200,1)).predict(advertising['TV'].reshape(200,1))
plt.scatter(advertising['TV'],advertising['Sales'])
plt.plot(advertising['TV'],y_,c='g')
```
