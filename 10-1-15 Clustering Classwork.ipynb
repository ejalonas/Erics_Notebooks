```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import cluster
```

```python
dentitio = pd.read_csv('https://raw.githubusercontent.com/galvin-mj/DAT_ATL_15/master/Datasets/dentitio.csv')
```

```python
dentitio.head()
```

```python
del dentitio['Unnamed: 0']
```

```python
x = dentitio.loc[:,'0':'7']
```

```python
x.head()
```

```python
z = cluster.hierarchy.linkage(x)
plt.figure(figsize=(50,20))
cluster.hierarchy.dendrogram(z, labels=dentitio['mammal'].values, orientation='right')
plt.show()
```
