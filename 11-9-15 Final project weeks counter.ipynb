```python
f= open('player_def_2015_1-8.pickle')
player_stat = pickle.load(f)
f.close()

g=open('team_def_2015_1-8.pickle')
team_stat = pickle.load(g)
g.close()

h=open('injuries_2015_1-8.pickle')
injuries_df = pickle.load(h)
h.close()
```

```python
for player in player_stat['name']:
    for week in player_stat[player_stat['name']==player]['week_num']:
        if player_stat[(injury_df['name_full']==player)&(injury_df['week_num']==week+1):
            player_stat[(player_stat['name']==player)&(player_stat['week_num']==week)]['injured?'] = 1
        else:
            player_stat[(player_stat['name']==player)&(player_stat['week_num']==week)]['injured?'] = 0
     
```

```python
player_stat[(player_stat['name']=='Erin Henderson')&(player_stat['week_num']==6)]['injured?']=1
```

```python
player_stat.reset_index(drop=True, inplace=True)
```

```python


for i in player_stat.index:
    name = player_stat.ix[i, 'name']
    week = player_stat.ix[i, 'week_num']
    if (len(injuries_df[(injuries_df['name_full']==name) & (injuries_df['week_num']==week+1)]) > 0):
        player_stat.loc[i,'injured?']=1
    #elif((len(injuries_df[(injuries_df['name_full']==name) & (injuries_df['week_num']==week)]) > 0)):
        #player_stat.loc[i,'injured?']=2
    else:
        player_stat.loc[i,'injured?']=0

```
